**Filter With More Than One Condition (AND)**

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

```python
mask1 = df["Gender"] == "Female" 
mask2 = df["Team"] == "Marketing"
df[ mask1 & mask2]
```

**Filter with More Than One Condition (OR)**

```python
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

```python
mask1 = df["Team"] == "Finance"
mask2 = df["Bonus %"] >= 5
df[ mask1 | mask2 ].head()
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Bonus %"] >= 5
mask3 = df["Team"] == "Finance"
df[ mask3 | (mask1 & mask2) ].head()
```

**The .isin() Method**

```python
import pandas as pd
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

**The .between() Method**

```python
import pandas as pd
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

```python
df["Salary"].between(60000, 120000) # Returns a series consisting of boolean values, True wherever value is in between range
df[ df["Salary"].between(60000, 120000) ]
```

```python
df["Bonus %"].between(6.0, 11.0)
df["Last Login Time"].between('8:00am', '2:30pm')
```

```python
df[ df["Bonus %"].between(6.0, 11.0) & df["Last Login Time"].between('8:00am', '2:30pm') ]
```

**The .duplicated() Method**

```python
import pandas as pd
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head()
```

```python
df.sort_values(by="First Name" ,inplace=True)
df["First Name"]
```

```python
df["First Name"].duplicated() 
df[ df["First Name"].duplicated() ]
```

```python
df["First Name"].duplicated(keep="first")
df[ df["First Name"].duplicated(keep="first") ] # This returns all rows which are duplicates
```

```python
df["First Name"].duplicated(keep="last")
```

```python
df["First Name"].duplicated(keep=False)  # Marks all occurences of any value duplicating as true
```

```python
~df["First Name"].duplicated(keep=False)
```

```python
df[~df["First Name"].duplicated(keep=False)]  # This will only return unique rows in the DataFrame
```

**The .drop_duplicates() Method**

```python
import pandas as pd
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace=True)
df.head()
```

```python
len(df)   # Returns the number of rows
```

```python
len(df.drop_duplicates()) # No rows have been dropped as no two rows are identical
```

```python
df.drop_duplicates(subset=["First Name"], keep="first") # Keeps first occurence of a repeating value and removes the rest
df.drop_duplicates(subset=["First Name"], keep="last")  # Keeps last occurence of a repeating value and removes the rest
df.drop_duplicates(subset=["First Name"], keep=False)  # Removes all instances of a repeating value
```

```python
df.drop_duplicates(subset=["First Name", "Team"]) # Subset parameter allow us to pass a list of columns to be considered while checking for duplicate values
```

**The .unique() and .nunique() Methods**

```python
import pandas as pd
df = pd.read_csv('employees.csv', parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace=True)
df.head()
```

```python
df["Gender"].unique()
```

```python
df["Team"].unique()
```

```python
len(df["Team"].unique()) # Returns the number of unique values in a series
```

```python
df["Team"].nunique() # Returns 10 as output instead of 11 above as it excludes NaN value by default
```

```python
df["Team"].nunique(dropna=False) # Returns NaN as unique values as well
```
