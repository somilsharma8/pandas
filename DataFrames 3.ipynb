**Retrieve Rows by Index Label with .loc[]**

```python
bond = pd.read_csv("jamesbond.csv", index_col="Film") # Setting Film as index to search through it later
bond.sort_index(inplace=True)
bond
```

```python
bond.loc["A View to a Kill"]  # Since we've passed only one index value, it returns a series
```

```python
bond.loc[["A View to a Kill", "GoldenEye"]] # Passing two index values returns a dataframe
```

```python
bond.loc["Diamonds Are Forever" : "Moonraker"] 
```

```python
bond.loc["Diamonds Are Forever" : ]  # Returns all rows lying after the provided index value, including the row at the provided index 
bond.loc[: "Moonraker"]  # Returns all rows lying up to the provided index value, including the row at the provided index 
```

```python
bond.loc[["Octopussy", "Moonraker", "Dr. No"]]
```

```python
bond.loc[["Moonraker", "Goldfinger", "GoldenEye", "Somil"]] 
# Passing a non-existenting index value in the list makes loc[] create a row with the new 
# index and NaN value in rest  of the columns
```

```python
"Somil" in bond.index # To check if a index value exists in the dataframe
```

**Set Secondary Values to a DataFrame**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
mask = bond["Actor"] == "Sean Connery"  # Storing filter in variable mask
bond[mask]    # Only fetches the rows where the filter is true
```

```python
df2 = bond[mask]
df2["Actor"] = "Sir Sean Connery"
df2
```

```python
bond
```

```python
mask = bond["Actor"] == "Sean Connery"      # create a filter
bond.loc[mask]    # This doesn't create a new dataframe made of filtered rows, it creates a view on the existing one
                 # Hence, any changes made to this will be permanent as it's the original dataframe only
```

```python
bond.loc[mask, "Actor"]  # Returns only the actor column for the rows filtered based on the mask value
```

```python
bond.loc[mask, "Actor"] = "Sir Sean Connery"   # Replacing all occurences of Sean Connery with sir Sean Connery
bond
```

**The .nsmallest() and .nlargest() Methods**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
# This is how we get top 3 largest value rows without the methods
bond.sort_values(['Box Office'], ascending=False).head(3)
```

```python
# This is how simplified it is to do the same thing with .nlargest()
bond.nlargest(3, 'Box Office')
bond.nlargest(3, ['Box Office', 'Budget'])
```

```python
# This is how simplified it is to do the same thing with .nsmallest()
bond.nsmallest(3, 'Box Office')
bond.nsmallest(3, ['Box Office', 'Budget'])
```

```python
# There's another format in which these methods can be used
bond['Box Office'].nlargest(3)
```

```python
bond['Year'].nsmallest(5)
```

**Filtering with the .where() Method**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
mask = bond['Actor'] == 'Sean Connery'

bond[mask]  # This would only return a subset of the current dataframe

bond.where(mask) # returns the dataframe itself, placing NaN in rows and columns that don't fulfill condition
```

```python
# Applying multiple conditions 
mask2 = bond['Bond Actor Salary'] > 2

bond.where(mask & mask2)
```

**Filtering with .query() Method**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
# Since our column names have spaces in them, we need to replace them first

bond.columns = [ column_names.replace(' ', '_') for column_names in bond.columns ] # Used square brackets as the for loop will return a list of updated column names
bond
```

```python
# Applying query method

bond.query('Actor == "Sean Connery"') # Note that input was passed as string, and condition within the input also has a string
```

```python
bond.query('Actor != "Sean Connery"')
```

```python
bond.query('Actor in ["Timothy Dalton", "Pierce Brosnan"] and Year > 1985')

bond.query('Actor not in ["Timothy Dalton", "Pierce Brosnan"] and Year > 1985')
```

**A Review of the .apply() method on Single Columns**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
def convert_to_string_and_add_millions(number):    # user defined function, the operation we wish to apply
    return str(number) + ' Millions!!'
```

```python
bond['Budget'].apply(convert_to_string_and_add_millions) # Returns a dataframe

bond['Budget'] = bond['Budget'].apply(convert_to_string_and_add_millions)
bond
```

**The apply() method with Row values**

```python
import pandas as pd
bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace=True)
bond
```

```python
def good_movie(row):
    actor = row[1]
    budget = row[4]
    
    if actor == 'Pierce Brosnan':
        return 'Best movie ever'
    
    elif actor == 'Roger Moore' and budget > 40.0:
        return 'Good movie to watch'
    else:
        return 'I have no idea'
    
bond.apply(good_movie, axis='columns')
```
