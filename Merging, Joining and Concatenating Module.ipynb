**The pd.concat() Method**

```python
week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
len(week1)
```

```python
len(week2)
```

```python
# The concatenation of week1 and week2 should result in a dataframe containing 500 rows in total

pd.concat([week1, week2]) # Here first 250 rows belong to week1 and the rest 250 belong to week2 because of the order in which
                            # they were passed
```

```python
len(pd.concat([week1, week2])) # Even though we have 500 rows in the resulting dataframe, the last index value is 249. Explained below
```

```python
pd.concat([week1, week2], ignore_index=True)
```

```python
pd.concat([week1, week2], keys=['Week 1', 'Week 2']) # Now we don't have to see by index number to check which dataframe the row belongs to
```

```python
sales = pd.concat([week1, week2], keys=['Week 1', 'Week 2'])
sales.loc['Week 1'] # Returns all the rows from concatenated dataframe which belong to Week 1
```

```python
sales.loc['Week 2'] # Returns all the rows from concatenated dataframe which belong to Week 1
```

```python
# If you want to fetch a record more specifically, we can pass a tupple into the loc[] method

sales.loc[('Week 1', 240)] # Passing both index values
```

```python
# If you now want to fetch a specific column value within that record, pass a column name apart from the tupple as well

sales.loc[('Week 1', 240), 'Customer ID']
```

**Outer Joins**

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
week = week1.merge(week2, how='outer', on='Customer ID', suffixes=[' - Week 1', ' - Week 2'])
week # Would only contain all the uncommon and common elements between the two dataframe
```

```python
week = week1.merge(week2, how='outer', on=['Customer ID'], indicator=True)
week
```

```python
# We can use .value_counts() emmthod to analyse where the data is coming from
week['_merge'].value_counts()
```

```python
mask = week['_merge'].isin(['left_only', 'right_only']) # To fetch only the uncommon records
week[mask]
```

**Left Join**

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
merged = week1.merge(foods, how='left', on='Food ID')
merged
```

```python
# Using the sort parameter sorts the records based on the column on which the two dataframes are being joined
merged = week1.merge(foods, how='left', on='Food ID', sort=True)
merged
```

**The left_on and right_on Parameters**

```python
import pandas as pd

week1 = pd.read_csv('Restaurant - Week 1 Sales.csv')
week2 = pd.read_csv('Restaurant - Week 2 Sales.csv')
customers = pd.read_csv('Restaurant - Customers.csv')
foods = pd.read_csv('Restaurant - Foods.csv')
```

```python
customers.head()
```

```python
week2.head()
```

```python
week2.merge(customers, how='left', left_on='Customer ID', right_on='ID')
```

```python
week2.merge(customers, how='left', left_on='Customer ID', right_on='ID').drop("ID", axis='columns') 
# We have to pass axis='columns' to drop() because it searches in axis='index' by default
```
