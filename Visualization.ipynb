**The .plot() Method**

```python
import pandas as pd
from pandas_datareader import data

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bb = data.DataReader(name="BB", data_source='yahoo', start='2007-07-01', end='2008-12-31')
bb.head()
```

```python
bb.plot()
```

```python
# Focussing on a single column for visualization

bb.plot(y = 'Volume')
```

```python
bb.plot(y='High')
```

```python
# To extract multiple columns for visualisation, pass a list of column names to the dataframe

bb[ ['High', 'Low'] ].plot()
```
