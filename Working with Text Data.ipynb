**The expand and n parameters of the .str.split() method**

```python
import pandas as pd
chicago = pd.read_csv('chicago.csv').dropna(how='all')
chicago['Department'] = chicago['Department'].astype('category')
chicago
```

```python
chicago['Name'].str.split(',', expand=True) # Returns a dataframe with first and last name stored in seperate columns
```

```python
# We can assign the above output to two newly created column in the chicago dataframe

chicago[['First Name', 'Last Name']] = chicago['Name'].str.split(',', expand=True)
chicago
```

```python
# Lets try and creating a dataframe by splitting the Position Title column

chicago['Position Title'].str.split(' ', expand=True)
```

```python
chicago['Position Title'].str.split(' ', n=1, expand=True)
```

```python
# Add the columns acquired above to the chicago dataframe

chicago[['First Title Word', 'Remaining Title Words']] = chicago['Position Title'].str.split(' ', expand=True, n=1)
chicago
```
