**Create a MultiIndex with the set_index() Method**

```python
import pandas as pd
bigmac = pd.read_csv('bigmac.csv', parse_dates=['Date'])
bigmac
```

```python
bigmac.set_index(keys='Country') # Single level index
```

```python
bigmac.set_index(keys='Date') # Single level index
```

```python
bigmac.set_index(keys=['Country', 'Date']) # MultiIndex dataframe
```

```python
bigmac.set_index(keys=['Date', 'Country']) # MultiIndex dataframe
```

```python
bigmac.set_index(keys=['Date', 'Country'], inplace=True)
bigmac
```

```python
bigmac.sort_index(inplace=True)
bigmac
```

```python
bigmac.index
```

```python
bigmac.index.names
```

```python
type(bigmac.index)
```

```python
bigmac.index[0]  # You can see how this returns a tupple or a record instead of a single value, so we need to pass two indices
                # to fetch a specific value
```

```python
bigmac.index[0][1]
```

**The .get_level_values() Method**

```python
import pandas as pd
bigmac = pd.read_csv('bigmac.csv', parse_dates=['Date'], index_col=['Date', 'Country']) # index_col has been used to create a multi
bigmac
```

```python
bigmac.index
```

```python
bigmac.index.get_level_values(0) # Returns all values in the first index column
bigmac.index.get_level_values(1) # Returns all values in the second index column
```
